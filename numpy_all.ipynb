# Boolean masking 

```python
import numpy as np

ar = np.array([12,34,36,43,45,23,56,78,57,43,12,45,78,8,45,89,87,9,65,32,32])

print(ar[ar >= 56])
 
```

```python
import numpy as np

ar = np.array([12,34,36,43,45,23,56,78,57,43,12,45,78,8,45,89,87,9,65,32,32])

print(ar[ar <= 65])
 
```

```python
import numpy as np

ar = np.array([12,34,36,43,45,23,56,78,57,43,12,45,78,8,45,89,87,9,65,32,32])

count =np.sum(ar>47)

print(count)
 
```

```python
a = np.array([2,4,8,3,9,1])

b = np.where(a % 2 == 0)

print(b)
 
```

```python
import numpy as np

ar = np.array([34,56,87,98,55,87,76,76,87,34,62,62,23,67,65,71,98])

mask =(ar> 45) & (ar < 79)

print(ar[mask])
 
```
